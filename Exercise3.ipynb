In this exercise, you will perform EDA on the popular Pokémon dataset, which contains detailed information about hundreds of Pokémon, including their stats, types, generation, and whether they are legendary.

The Pokémon dataset is ideal for practicing EDA because it includes a mix of numerical features (such as Attack, Defense, Speed, HP) and categorical features (such as Type 1, Type 2, and Generation). This combination allows you to apply a wide range of EDA techniques, including summary statistics, data visualizations, grouping and aggregation, correlation analysis, and comparisons across categories.

Throughout the exercise, you will explore questions such as:
- Which Pokémon tend to have the highest or lowest stats?
- How do different Pokémon types compare in terms of strength or defense?
- Do Legendary Pokémon differ significantly from non-Legendary ones?
- How are various stats distributed across the entire dataset?
- Are there relationships or trade-offs between certain attributes (e.g., Attack vs. Defense)?

## 2. Summary Statistics (4 pts)

1. Generate `df.describe()`.

```python
df.describe()
```

2. Get mean, median, and mode of Attack.

```python
mean_attack = df['Attack'].mean()
median_attack = df['Attack'].median()
mode_attack = df['Attack'].mode()

print(f"Mean Attack: {mean_attack}")
print(f"Median Attack: {median_attack}")
print(f"Mode Attack:\n{mode_attack}")
```

3. Compute 25th and 75th percentiles for HP.

```python
hp_25th_percentile = df['HP'].quantile(0.25)
hp_75th_percentile = df['HP'].quantile(0.75)

print(f"25th percentile for HP: {hp_25th_percentile}")
print(f"75th percentile for HP: {hp_75th_percentile}")
```

4. Compute standard deviation and variance of Speed.

```python
std_speed = df['Speed'].std()
variance_speed = df['Speed'].var()

print(f"Standard Deviation of Speed: {std_speed}")
print(f"Variance of Speed: {variance_speed}")
```

## 4. Categorical Exploration `(9 pts)`

Find the number of Pokémon per primary type.

```python
df['Type 1'].value_counts()
```

Find the number of Pokémon per generation.

```python
df['Generation'].value_counts()
```

Which type appears the most? Which appears the least?

```python
all_types = pd.concat([df['Type 1'], df['Type 2']]).dropna()

most_common_type = all_types.value_counts().idxmax()
least_common_type = all_types.value_counts().idxmin()

print(f"The most common type (including Type 1 and Type 2) is: {most_common_type}")
print(f"The least common type (including Type 1 and Type 2) is: {least_common_type}")
```

How many unique primary types (Type 1) exist?

```python
unique_primary_types = df['Type 1'].nunique()
print(f"Number of unique primary types (Type 1): {unique_primary_types}")
```

```python
unique_primary_types_list = df['Type 1'].unique()
print(f"Unique Primary Types (Type 1): {unique_primary_types_list}")
```

How many unique secondary types (Type 2) exist?

```python
unique_secondary_types = df['Type 2'].nunique()
print(f"Number of unique secondary types (Type 2): {unique_secondary_types}")
```

```python
unique_secondary_types_list = df['Type 2'].unique()
print(f"Unique Secondary Types (Type 2): {unique_secondary_types_list}")
```


Which primary types have the most dual-type combinations?

```python
df.groupby(['Type 1', 'Type 2']).size().sort_values(ascending=False)
```

Which type has the highest mean Attack?

```python
df.groupby('Type 1')['Attack'].mean().idxmax()
```

Which type has the lowest mean Defense?


Which generation has the highest average Speed?

```python
df.groupby('Generation')['Speed'].mean().idxmax()
```

## 5. Groupby & Aggregation `(13 pts)`

Compute the average Attack per primary type.

```python
average_attack_per_type = df.groupby('Type 1')['Attack'].mean()
print(average_attack_per_type)
```

Compute the maximum HP per generation.


```python
max_hp_per_generation = df.groupby('Generation')['HP'].max()
print(max_hp_per_generation)
```

Compute the total number of Pokémon per primary type.

```python
total_pokemon_per_type = df['Type 1'].value_counts()
print(total_pokemon_per_type)
```

For each type, compute:

- mean Attack

- mean Defense

- mean Speed

```python
mean_stats_per_type = df.groupby('Type 1')[['Attack', 'Defense', 'Speed']].mean()
print(mean_stats_per_type)
```

For each generation, compute:

- count

- mean Total

- number of Legendary Pokémon (hint: use sum on Boolean)

```python
generation_stats = df.groupby('Generation').agg(
    count=('Name', 'size'),
    mean_total=('Total', 'mean'),
    num_legendary=('Legendary', 'sum')
)
print(generation_stats)
```

Which type combination (Type 1 + Type 2) has the highest average Attack?

```python
highest_avg_attack_combo = df.groupby(['Type 1', 'Type 2'])['Attack'].mean().idxmax()
print(f"The type combination with the highest average Attack is: {highest_avg_attack_combo}")
```

Which generation has the highest proportion of Legendary Pokémon?

```python
highest_num_legendary_gen = generation_stats['num_legendary'].idxmax()
print(f"Generation {highest_num_legendary_gen}")
```


Which primary type has the largest variance in HP?

```python
largest_hp_variance_type = df.groupby('Type 1')['HP'].var().idxmax()
print(f"The primary type with the largest variance in HP is: {largest_hp_variance_type}")
```

Which primary type has the highest median Speed?

```python
highest_median_speed_type = df.groupby('Type 1')['Speed'].median().idxmax()
print(f"The primary type with the highest median Speed is: {highest_median_speed_type}")
```

Group Pokémon by whether they are Legendary or not. Compare:

- mean Total

- mean Attack

- mean Defense

- mean Speed

```python
legendary_stats_comparison = df.groupby('Legendary')[['Total', 'Attack', 'Defense', 'Speed']].mean()
print(legendary_stats_comparison)
```

Show the top 5 strongest types by mean Total.

```python
top_5_strongest_types = df.groupby('Type 1')['Total'].mean().nlargest(5)
print(top_5_strongest_types)
```

Rank generations by their average Attack.

```python
avg_attack_per_generation = df.groupby('Generation')['Attack'].mean().sort_values(ascending=False)
print("Generations ranked by average Attack:\n", avg_attack_per_generation)
```

Show the top 10 fastest Pokémon using nlargest(10, "Speed").

```python
top_10_fastest_pokemon = df.nlargest(10, "Speed")
print(top_10_fastest_pokemon)
```
